```python
path_to_annotations = [f"{HOME}/train/labels/_annotations.coco.json",f"{HOME}/test/labels/_annotations.coco.json" ,f"{HOME}/valid/labels/_annotations.coco.json"]
path_to_images = [f"{HOME}/train/images",f"{HOME}/test/images" ,f"{HOME}/valid/images" ]
path_to_output = [f"{HOME}/train/labels",f"{HOME}/test/labels",f"{HOME}/valid/labels"]


for i in range(len(path_to_annotations)):
  dataset = importer.ImportCoco(path_to_annotations[i],path_to_images=path_to_images[i],name="COCO_An")
  dataset.export.ExportToYoloV5(path_to_output[i],"data.yaml",use_splits=True,segmentation=True)



```
